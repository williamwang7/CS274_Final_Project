```python
import numpy as np
import cv2

with open('/content/gdrive/MyDrive/CUTE80_Dataset/Groundtruth/GroundTruth.xml', 'r') as file:
  data = file.readlines()
file.close()

with open('/content/gdrive/MyDrive/scrabble-gan/gt_cute80.txt', 'r') as file:
  data_gt_cute = file.readlines()
file.close()

img_and_coords = []
counter = 0
while counter < len(data):
  if data[counter].find('<ImageName>') != -1:
    start = data[counter].find('<ImageName>') + 11
    end = data[counter].find('</ImageName>')
    img_and_coords.append(data[counter][start:end])
  elif data[counter].find('<PolygonPoints ') != -1:
    coords = []
    start_x = data[counter].find('x=') + 2
    start_y = data[counter].find('y=') + 2
    string_x_coords = data[counter][start_x + 1:data[counter].find('"', start_x + 1)].split()
    string_y_coords = data[counter][start_y + 1:data[counter].find('"', start_y + 1)].split()
    for i in range(0, len(string_x_coords)):
      coords.append([int(string_x_coords[i]), int(string_y_coords[i])])
    img_and_coords.append(coords)
  counter += 1

cur_img = None
cur_img_name = ""
text = ""
counter = 0
for i in range(0, len(img_and_coords)):
  if isinstance(img_and_coords[i], str):
    cur_img_name = img_and_coords[i]
    cur_img = cv2.imread('/content/gdrive/MyDrive/CUTE80_Dataset/CUTE80/' + cur_img_name)
  elif isinstance(img_and_coords[i], list):
    if cur_img is not None:
      x, y, w, h = cv2.boundingRect(np.array(img_and_coords[i]))
      cropped_img = cur_img[y:y + h, x:x + w].copy()
      if cropped_img.size != 0:
        cv2.imwrite('/content/gdrive/MyDrive/scrabble-gan/data/CUTE80/words/img_' + str(counter) + '.png', cropped_img)
        if data_gt_cute[counter][0] == '*':
          text += "img_" + str(counter) + " err 154 " + str(x) + " " + str(y) + " " + str(w) + " " + str(h) + " CUTE80 " + data_gt_cute[counter][1:] + "\n"
        else:
          text += "img_" + str(counter) + " ok 154 " + str(x) + " " + str(y) + " " + str(w) + " " + str(h) + " CUTE80 " + data_gt_cute[counter] + "\n"
        counter += 1

with open('/content/gdrive/MyDrive/scrabble-gan/data/CUTE80/words.txt', 'w') as file:
  file.write(text)
file.close()
```

```python
with open('/content/gdrive/MyDrive/scrabble-gan/coords_idcar2015.txt', 'r') as file:
  data_coords = file.readlines()
file.close()

with open('/content/gdrive/MyDrive/scrabble-gan/gt_idcar2015.txt', 'r') as file:
  data_gt = file.readlines()
file.close()

text = ""
for i in range(0, len(data_coords)):
  start_coords = data_coords[i].find(',') + 1
  string_coords = data_coords[i][start_coords:-1].split(",")
  for j in range(0, len(string_coords)):
    string_coords[j] = int(string_coords[j])
  coords = np.array(string_coords).reshape(int(len(string_coords)/2), 2)
  x, y, w, h = cv2.boundingRect(coords)
  start_gt = data_gt[i].find('"') + 1
  word = data_gt[i][start_gt:-2].replace(" ", "").replace("É", "E")
  if len(word) <= 10:
    text += data_coords[i][:start_coords - 5] + " ok 154 " + str(x) + " " + str(y) + " " + str(w) + " " + str(h) + " IDCAR2015 " + word + "\n"

with open('/content/gdrive/MyDrive/scrabble-gan/data/IDCAR2015/words.txt', 'w') as file:
  file.write(text)
file.close()
```
